```python
import sqlite3
import pandas as pd
```

```python
%load_ext sql
```

```python
con = sqlite3.connect('spacex.db')
cur = con.cursor()
```

```python
%sql sqlite:///spacex.db
```

```python
df = pd.read_csv('SpaceX_f9_clean.csv')
```

```python
df.to_sql('spacex_clean',con, index = False,if_exists = 'replace')
```

```python
%%sql
select count(*) from spacex_clean ;
```

```python
%%sql
select * from spacex_clean
limit 5
```

```python
%%sql
select distinct(launchsite) from spacex_clean
```

```python
%%sql
select * from spacex_clean
where launchsite like 'CCS%'
limit 5
```

```python
%%sql
select distinct(BoosterVersion) from spacex_clean
```

```python
%%sql
select min(date) from spacex_clean
where class = 1
```

```python
%%sql
select min(date) from spacex_clean
where class = 1 
```

```python
%%sql
select distinct(outcome) from spacex_clean
```

```python
%%sql
select count(*), class from spacex_clean
group by class
```

```python
%%sql
select count(*), outcome from spacex_clean
group by outcome
```

```python
%%sql
select max(payloadmass) from spacex_clean
```

```python
%%sql
select * from spacex_clean
where payloadmass =( select max(payloadmass) from spacex_clean)
```

```python
%%sql
select outcome, count(outcome) as Number from spacex_clean
where date between '2010-06-04' and '2017-03-20' 
group by outcome
```
